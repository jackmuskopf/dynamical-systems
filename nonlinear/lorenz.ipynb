```python
import numpy as np
import scipy.integrate
import matplotlib.pyplot as plt
```

```python
sigma = 10
beta = 8/3
rho = 28
```

```python
# dynamics
def xdot(x, t):
    x,y,z = x
    return np.array([
        sigma*(y - x),
        x*(rho - z) - y,
        x*y - beta*z
    ])
```

```python
x0 = [1.0, 1.0, 1.0]
t = np.linspace(0.0, 40.0, 1200)
```

```python
xt = scipy.integrate.odeint(xdot, x0, t)
```

```python
fig = plt.figure()
ax = fig.gca(projection="3d")
ax.plot(xt[:, 0], xt[:, 1], xt[:, 2])
plt.draw()
plt.show()
```
